```python
n = 9
m = n // 2

# Lists to store each letter's rows
I, L, O, V, E, U = [], [], [], [], [], []

# I
for i in range(n):
    row = ""
    for j in range(n):
        if i == 0 or i == n - 1 or j == m:
            row += "* "
        else:
            row += "  "
    I.append(row)

# L
for i in range(n):
    row = ""
    for j in range(n):
        if i == n - 1 or j == 0:
            row += "* "
        else:
            row += "  "
    L.append(row)

# O
for i in range(n):
    row = ""
    for j in range(n):
        if i == 0 or i == n - 1 or j == 0 or j == n - 1:
            row += "* "
        else:
            row += "  "
    O.append(row)

# V
for i in range(n):
    row = ""
    for j in range(n * 2 - 1):
        if j == i or j == (2 * n - 2 - i):
            row += "*"
        else:
            row += " "
    V.append(row)

# E
for i in range(n):
    row = ""
    for j in range(n):
        if i == 0 or i == m or i == n - 1 or j == 0:
            row += "* "
        else:
            row += "  "
    E.append(row)

# U
for i in range(n):
    row = ""
    for j in range(n):
        if i == n - 1 or j == 0 or j == n - 1:
            row += "* "
        else:
            row += "  "
    U.append(row)

# Print side-by-side
for i in range(n):
    print(I[i] + "  " + L[i] + "  " + O[i] + "  " + V[i] + "  " + E[i] + "  " + U[i])

```

```python


B, U, J1, J2, I = [], [], [], [], []

# Letter B
n=5
m=n//2
for i in range(n):
    row = ""
    for j in range(n):
        if i == n-1 or j == 0 or i==0 or i==m or j==n-1:
            row += "* "
        else:
            row += "  "
    B.append(row)

# Letter U
for i in range(n):
    row = ""
    for j in range(n):
        if (j == 0 or j == n-1) and i != n-1:
            row += "* "
        elif i == n-1 and j != 0 and j != n-1:
            row += "* "
        else:
            row += "  "
    U.append(row)
    
# letter j
for i in range(n):
    row=" "
    for j in range(n):
        if j == n-1 or i==n-1  :
            row+="*"
        else:
            row+=" "
    J1.append(row)
    

# letter j2    
for i in range(n):
    row=" "
    for j in range(n):
        if j == n-1 or i==n-1  :
            row+="*"
        else:
            row+=" "
    J2.append(row)



# Letter I
for i in range(n):
    row = ""
    for j in range(n):
        if  i ==0 or i==n-1 or j==m:
            row += "* "
        else:
            row += "  "
    I.append(row)

# Print all letters together
for i in range(n):
    print(B[i] + "   " + U[i] + "   " + J1[i] + "   " + J2[i] + "   " + I[i])
```
